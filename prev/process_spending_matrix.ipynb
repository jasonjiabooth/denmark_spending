# Part 1: Convert spending matrix from wide to long

```python
df = pd.read_csv(inputfilepath)
```

```python
df.rename(columns = {'cust_muni':'spender_region', 'spend_muni': 'recipient_region', 
                     'cust_sector':'spender_sector','spend_sector': 'recipient_sector',
                     'cust_cell':'spender_sector_region', 'amount_dkk': 'spend'}, inplace = True)
```

```python
df.head()
```

```python
id_vars = ['spender_region', 'spender_sector', 'spender_sector_region']
```

```python
df_new = df
```

# Code to execute if the variable 'recipient_sector_region' exists

```python
df_new = pd.melt(df, id_vars=id_vars, var_name='recipient_sector_region', value_name='spend')
df_new
```

```python
df_new['recipient_region'] = df_new['recipient_sector_region'].str.extract(r'spendinflow_(.+)$')
df_new['recipient_sector'] = df_new['recipient_sector_region'].str.extract(r'(.+)_.+$')
df_new
```

```python
df_new[df_new['spender_sector_region']=='s24_nospendinflow_foreign'] # Test
```

```python
df_new['spender_sector'] = df_new['spender_sector'].str[1:3]
df_new['recipient_sector'] = df_new['recipient_sector'].str[3:5]
df_new
```

```python
df_new = df_new.reindex(columns=['spender_region', 'spender_sector', 'recipient_region', 'recipient_sector',
                                 'spender_sector_region', 'recipient_sector_region', 'spend'])
df_new
```

# Part 2: Convert region codes into actual region names

```python
region_names = pd.read_excel(r"C:\Users\jasonjia\Dropbox\Projects\denmark_spending\data\dk_municipalities_code_to_name.xlsx")
```

```python
region_names.shape
```

```python
region_names.head()
```

```python
# region_names = region_names.append({'Code': 'foreign', 'Name': 'Foreign'}, ignore_index=True) # Already done so in the Excel sheet
```

```python
region_names
```

```python
# Because region codes contain numbers (e.g. 101) and strings (i.e. 'foreign'), 
# we need to convert them both to the same data type.
region_names['Code'].dtype
```

```python
region_names['Code'] = region_names['Code'].astype('str')
region_names['Code'].dtype
```

```python
df_new['spender_region'] = df_new['spender_region'].astype('str')
df_new['spender_region'].dtype
```

```python
df_new_names = df_new.merge(region_names, left_on='spender_region', right_on='Code', how='left')
df_new_names
```

```python
df_new_names = df_new_names.rename(columns = {'Name':'spender_region_name'})
df_new_names = df_new_names.drop(['Code'], axis = 1)
df_new_names
```

```python
df_new_names = df_new_names.merge(region_names, left_on='recipient_region', right_on='Code', how='left')
df_new_names
```

```python
df_new_names = df_new_names.rename(columns = {'Name':'recipient_region_name'})
df_new_names = df_new_names.drop(['Code'], axis = 1)
df_new_names
```

```python
df_new_names = df_new_names.reindex(columns=['spender_region', 'spender_region_name', 'spender_sector', 
                                             'recipient_region', 'recipient_region_name', 'recipient_sector',
                                             'spend'])
df_new_names
```
